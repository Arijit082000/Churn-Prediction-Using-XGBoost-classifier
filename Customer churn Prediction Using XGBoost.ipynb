```python
#Customer ID is not necessary
df.drop('customerID', axis =1, inplace=True)
```

```python
#Convert target Churn columns values yes, no to 1,0
df['Churn'] = df['Churn'].map({'Yes':1,'No':0})
```

```python
#Encode all categorical columns by one hot encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
#Find only Boolean columns from dataframe
bool_col = df.select_dtypes(include=['bool']).columns
#Convert only Boolean columns to integer
df[bool_col] = df[bool_col].astype(int)
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
```

```python
#Separate feature (x) and target (y)
x = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
#Split train test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python

# Class 0 এবং Class 1 এর অনুপাত বের করা
ratio = float(np.sum(y_train == 0)) / np.sum(y_train == 1)

#Create XGboost model and fit
model = XGBClassifier(
n_estimators = 100,
learning_rate=0.1,
max_depth=5,
scale_pos_weight=ratio,  # এই প্যারামিটারটি Class 1 এর প্রতি ফোকাস বাড়াবে

random_state=42)

model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

pred = model.predict(x_test)
```

```python
print("Accuracy Score: ", accuracy_score(y_test, pred))
print("Classification Report: ")
print(classification_report(y_test, pred))
```

```python
#Ploting Confusion Matrix
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, pred), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
import xgboost as xgb
plt.figure(figsize=(10,8))
xgb.plot_importance(model, importance_type='weight', max_num_features=10)
plt.title('Top 10 Most Important Features for Customer Churn')
plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd

# 1. Creating all input fields (widgets)
print("--- Customer Churn Predictor ---")

# Demographic data
gender_dd = widgets.Dropdown(options=['Female', 'Male'], description='Gender:')
senior_dd = widgets.Dropdown(options=['No', 'Yes'], description='SeniorCitizen:')
partner_dd = widgets.Dropdown(options=['Yes', 'No'], description='Partner:')
dependents_dd = widgets.Dropdown(options=['Yes', 'No'], description='Dependents:')

# Service Data
tenure_slider = widgets.IntSlider(min=0, max=100, value=1, description='Tenure (Months):')
phone_dd = widgets.Dropdown(options=['Yes', 'No'], description='PhoneService:')
multiple_lines_dd = widgets.Dropdown(options=['Yes', 'No', 'No phone service'], description='MultipleLines:')
internet_dd = widgets.Dropdown(options=['DSL', 'Fiber optic', 'No'], description='Internet:')
security_dd = widgets.Dropdown(options=['Yes', 'No', 'No internet service'], description='OnlineSecurity:')
backup_dd = widgets.Dropdown(options=['Yes', 'No', 'No internet service'], description='OnlineBackup:')
device_dd = widgets.Dropdown(options=['Yes', 'No', 'No internet service'], description='DeviceProtect:')
support_dd = widgets.Dropdown(options=['Yes', 'No', 'No internet service'], description='TechSupport:')
tv_dd = widgets.Dropdown(options=['Yes', 'No', 'No internet service'], description='StreamingTV:')
movies_dd = widgets.Dropdown(options=['Yes', 'No', 'No internet service'], description='StreamingMovies:')

# Payment Data
contract_dd = widgets.Dropdown(options=['Month-to-month', 'One year', 'Two year'], description='Contract:')
paperless_dd = widgets.Dropdown(options=['Yes', 'No'], description='PaperlessBill:')
payment_dd = widgets.Dropdown(options=['Electronic check', 'Mailed check', 'Bank transfer (automatic)', 'Credit card (automatic)'], description='Payment:')
monthly_charges_input = widgets.FloatText(value=50.0, description='MonthlyCharges:')
total_charges_input = widgets.FloatText(value=50.0, description='TotalCharges:')

# Button and Output
predict_button = widgets.Button(description="Predict Churn", button_style='success')
output = widgets.Output()

# 2. Prediction logic when the button is clicked
def on_predict_button_clicked(b):
    with output:
        clear_output()

        # Create a dictionary from user input
        input_data = {
            'gender': [gender_dd.value],
            'SeniorCitizen': [senior_dd.value],
            'Partner': [partner_dd.value],
            'Dependents': [dependents_dd.value],
            'tenure': [tenure_slider.value],
            'PhoneService': [phone_dd.value],
            'MultipleLines': [multiple_lines_dd.value],
            'InternetService': [internet_dd.value],
            'OnlineSecurity': [security_dd.value],
            'OnlineBackup': [backup_dd.value],
            'DeviceProtection': [device_dd.value],
            'TechSupport': [support_dd.value],
            'StreamingTV': [tv_dd.value],
            'StreamingMovies': [movies_dd.value],
            'Contract': [contract_dd.value],
            'PaperlessBilling': [paperless_dd.value],
            'PaymentMethod': [payment_dd.value],
            'MonthlyCharges': [monthly_charges_input.value],
            'TotalCharges': [total_charges_input.value]
        }

        input_df = pd.DataFrame(input_data)

        # Converting text data into numbers (One-Hot Encoding)
        input_encoded = pd.get_dummies(input_df)

        try:
            # [Most important step]: Match the input columns exactly with the model's training features
            # Here, 'model' is the variable name of your XGBoost model
            training_features = model.feature_names_in_
            input_final = input_encoded.reindex(columns=training_features, fill_value=0)

            # Making a prediction
            prediction = model.predict(input_final)

            print("\n==================================")
            if prediction[0] == 1:
                print(" Result: Yes (there is a chance the customer might leave the service)")
            else:
                print(" Result: No (the customer will continue)")
            print("==================================")

        except Exception as e:
            print(f"Error: {e}")
            print("Please make sure your model's name is 'model' and that it is running in the cell above.")

predict_button.on_click(on_predict_button_clicked)

# 3. Arrange the form neatly in 3 columns and show it in a collapsible way
vbox1 = widgets.VBox([gender_dd, senior_dd, partner_dd, dependents_dd, tenure_slider, phone_dd, multiple_lines_dd])
vbox2 = widgets.VBox([internet_dd, security_dd, backup_dd, device_dd, support_dd, tv_dd, movies_dd])
vbox3 = widgets.VBox([contract_dd, paperless_dd, payment_dd, monthly_charges_input, total_charges_input])

ui = widgets.HBox([vbox1, vbox2, vbox3])
display(ui, predict_button, output)
```
